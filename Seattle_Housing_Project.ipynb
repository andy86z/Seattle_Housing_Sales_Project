```python
import pandas as pd
```

```python
housing_data = pd.read_csv('kc_house_data.csv')
housing_data.head()
```

```python
housing_data.columns
```

```python
import datetime
current_year = datetime.datetime.now().year
print(current_year)
```

```python
housing_data['age_of_house'] = current_year - pd.to_datetime(housing_data['date']).dt.year
housing_data.head()
```

```python
housing_data.info()
```

```python
housing_data.columns
```

```python
feature_cols = [ 'age_of_house',  'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']
```

```python
x = housing_data[feature_cols]
x.head()
```

```python
y = housing_data['price']
y.head()
```

```python
import seaborn as sns
%matplotlib inline

sns.pairplot(housing_data, x_vars=feature_cols, y_vars='price',size=7,aspect=0.7, kind = 'reg')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
accuracy = regressor.score(x_test,y_test)
f"Accuracy: {int(round(accuracy * 100))}%"
```
